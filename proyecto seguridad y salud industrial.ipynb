### Preprocesamiento de NLP

  De la columna de texto 'Descripcion' se va convertir todo el texto a minúsculas para unificarlo, aplicar lematización o stemming para reducir las palabras a su forma raíz y eliminar palabras comunes sin valor informativo, como artículos y conectores

```python
# Ademas de las palabras de WORDCLOUD se agregara otras palabras clave para el analisis
STOPWORDS.update(["cm", "kg", "mr", "wa" ,"nv", "ore", "da", "pm", "am", "cx"])
print(STOPWORDS)
```

```python
def nlp_preprocesser(row):
    sentence = row.Description
    #convitiendo todo a minusculas
    lowered = sentence.lower()
    tok = tokenize.word_tokenize(lowered)

    #Reduce las palabras a su forma base (lema)
    #eliminando stopwords
    lemmatizer = stem.WordNetLemmatizer()
    lem = [lemmatizer.lemmatize(i) for i in tok if i not in STOPWORDS]
    stemmer = stem.PorterStemmer()
    stems = [stemmer.stem(i) for i in lem if i not in STOPWORDS]

    #elminina caracteres alfanumericos como '(', '.' or '!'
    alphas = [i for i in stems if i.isalpha() and (i not in STOPWORDS)]
    return " ".join(alphas)
```

```python
#visualizacion de nueva columna Description_processed

df['Description_processed'] = df.apply(nlp_preprocesser, axis=1)
df.head(3)
```

### Analisis de Sentimiento





El análisis de sentimiento con VADER se aplicará a las descripciones de los accidentes para identificar la polaridad del texto (positivo, negativo o neutro). Esto permitirá extraer información sobre el tono emocional asociado a cada accidente, complementando el análisis de palabras clave.



```python
def sentiment2score(text):
    analyzer = SentimentIntensityAnalyzer()
    sent_score = analyzer.polarity_scores(text)["compound"]
    return float(sent_score)
```

```python
df['Description_sentiment_score'] = df['Description'].apply(lambda x: sentiment2score(x))
df.head(3)
#+1: Indica un sentimiento extremadamente positivo. Un texto con una puntuación de +1 se considera que expresa una emoción o actitud muy favorable.
#-1: Indica un sentimiento extremadamente negativo. Un texto con una puntuación de -1 se considera que expresa una emoción o actitud muy desfavorable.
```

### ANALISIS NLP

La descripción de los accidentes es fundamental para entender sus causas, por lo que es necesario identificar palabras o frases clave que reflejen las circunstancias en las que se produjeron los incidentes.

```python
def ngram_func(ngram, trg='', trg_value=''):
    #trg_value is list-object
    if (trg == '') or (trg_value == ''):
        string_filterd =  df['Description_processed'].sum().split()
    else:
        string_filterd =  df[df[trg].isin(trg_value)]['Description_processed'].sum().split()
    dic = nltk.FreqDist(nltk.ngrams(string_filterd, ngram)).most_common(30)
    ngram_df = pd.DataFrame(dic, columns=['ngram','count'])
    ngram_df.index = [' '.join(i) for i in ngram_df.ngram]
    ngram_df.drop('ngram',axis=1, inplace=True)
    return ngram_df

hv.Bars(ngram_func(1)[::-1]).opts(title="Palabra top 30", color="green", xlabel="Unigrams", ylabel="Count")\
                .opts(opts.Bars(width=600, height=600,tools=['hover'],show_grid=True,invert_axes=True))
```

```python
hv.Bars(ngram_func(2)[::-1]).opts(title="2 Palabras top 30", color="yellow", xlabel="Bigrams", ylabel="Count")\
                .opts(opts.Bars(width=600, height=600,tools=['hover'],show_grid=True,invert_axes=True))
```

```python
hv.Bars(ngram_func(3)[::-1]).opts(title="3 Palabras top 30", color="blue", xlabel="Trigrams", ylabel="Count")\
                .opts(opts.Bars(width=600, height=600,tools=['hover'],show_grid=True,invert_axes=True))
```
